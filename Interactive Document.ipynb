<font face="courier" color="black">
<h2>
    <i>
        Dealing with Outliers
    </i>
</h2>
<br>
Ok, so we (unfortunately) can't just arbitrarily choose which data points we think are outliers. Why? Well, sure, the first few data points might be really obviously out of the way, so getting rid of them is fine. But when do we stop? Every time we take out a data point that is far away from the others, we make out dataset smaller. <i>Why is this a problem?</i> you might sheepishly ask. <br><br>

Well, of course, you and I have only the best intentions. However, imagine a more ~*nefarious*~ researcher, wanting to make sure their data shows statistical importance because they need to publish a paper to keep their job, or to get more grant money, or what have you. They might keep eliminating "outliers" because the more of them they leave out of their dataset, the stronger their statistical relevance tests get. This is unethical, and while I could go into more detail about why it is unethical, that's not part of our journey today.<br><br>

Luckily, we have computer science to remove the outliers for us! The visualization below will be similar to the one you just saw, but it will now flag the data points that are outliers by coloring them in a dark periwinkle-ish purple.

</font>

```python
# Catch all bokeh warnings from plotting different sized dfs
import warnings
warnings.filterwarnings('ignore') 

state1 = {
    'x_change': 'automated readability index',
    'y_change': 'coleman liau index',
}

# FIND X AND Y DATA, PASS TO UPDATE_NOTEBOOK()
def create_plot1(state1):
    x_data = reduced_dataset[state1['x_change']].values.T.tolist()
    y_data = reduced_dataset[state1['y_change']].values.T.tolist()
    return x_data, y_data
    
# Helper method that creates a boolean array of whether (true) or not (false) the point is an outlier
def outliers(data):
    outliers = []
    booleans = [0]*len(data)
    mean = sum(data)/(1.0*len(data))
    var = sum((data[i] - mean)**2 for i in range(0, len(data)))/(1.0*len(data))
    std = var**0.5
    outliers = [data[i] for i in range(0, len(data)) if abs(data[i]-mean) > 1.96*std]
    for datum in data:
        if datum in outliers:
            index = data.index(datum)
            booleans[index] = 1
    return booleans        
    

def full_data(plot):
    x_data = plot[0]
    y_data = plot[1]
    full_data_x = []
    full_data_y = []
    for i in range(len(x_data)):
        full_data_x.append(x_data[i])
        full_data_y.append(y_data[i]) 
    return full_data_x, full_data_y

def outlier_indices(plot, outliers_x, outliers_y):
    x_data = plot[0]
    y_data = plot[1]
    outlier_ind = []
    for i in range(len(x_data)):
        if outliers_x[i] or outliers_y[i]:
            outlier_ind.append(i)
    return outlier_ind

def regular_outlier_data(plot, outlier_ind, full_dataset):
    x_data = plot[0]
    y_data = plot[1]
    full_x = full_dataset[0]
    full_y = full_dataset[1]
    outlier_data_x = []
    outlier_data_y = []
    regular_data_x = []
    regular_data_y = []
    for i in range(len(x_data)):
        if i in outlier_ind:
            outlier_data_x.append(full_x[i])
            outlier_data_y.append(full_y[i])
        else:
            regular_data_x.append(full_x[i])
            regular_data_y.append(full_y[i])
    outlier_data = [outlier_data_x, outlier_data_y]
    regular_data = [regular_data_x, regular_data_y]
    return outlier_data, regular_data

def separate_data(plot, outliers_x, outliers_y):
    full_dataset = full_data(plot)
    outlier_ind = outlier_indices(plot, outliers_x, outliers_y)
    all_data = regular_outlier_data(plot, outlier_ind, full_dataset)
    return all_data

# GET DATA
plot1 = create_plot1(state1) 
outliers_x = outliers(plot1[0]) 
outliers_y = outliers(plot1[1])
all_data = separate_data(plot1, outliers_x, outliers_y)
outlier_data = all_data[0]
regular_data = all_data[1]

source_regular = ColumnDataSource(
    data=dict(
        x_reg=regular_data[0],
        y_reg=regular_data[1],
        desc=titles, 
))

source_outliers = ColumnDataSource(
    data=dict(
         x_out=outlier_data[0],
         y_out=outlier_data[1],
         desc=titles,
))

hover1 = HoverTool(tooltips=[
    ("(x,y)", "($x, $y)"),
    ("desc", "@desc"),
])

second_figure = figure(plot_width=400, plot_height=400, tools=[hover1],
           title="Correlation Plot: Managing Outliers")

renderer_regular = second_figure.circle(
         'x_reg', 
         'y_reg', 
         size=8, 
         color='darkmagenta',
         alpha=0.6,
         legend="regular data",
         source=source_regular)

renderer_outliers = second_figure.circle(
        'x_out',
        'y_out',
        size=8,
        color='mediumslateblue',
        alpha=0.6,
        legend="outlier data",
        source=source_outliers
        )

second_figure.xaxis.axis_label = state1['x_change']
second_figure.yaxis.axis_label = state1['y_change']

show(second_figure, notebook_handle=True)


# PUSH CHANGES TO NOTEBOOK
def update_notebook1():
    x, y = create_plot1(state1)
    plot = [x, y]
    outliers_x = outliers(plot[0])
    outliers_y = outliers(plot[1])
    all_data = separate_data(plot, outliers_x, outliers_y)
    outlier_data = all_data[0]
    regular_data = all_data[1]
    renderer_regular.data_source.data['x_reg'] = regular_data[0]
    renderer_regular.data_source.data['y_reg'] = regular_data[1]
    renderer_outliers.data_source.data['x_out'] = outlier_data[0]
    renderer_outliers.data_source.data['y_out'] = outlier_data[1]
    second_figure.xaxis.axis_label = state1['x_change']
    second_figure.yaxis.axis_label = state1['y_change']
    push_notebook()

# TRACK CHANGES TO X AND Y FROM DROPDOWNS
def change_x1(change):
    global state1
    state1['x_change'] = change.new
    update_notebook1()
    
def change_y1(change):
    global state1
    state1['y_change'] = change.new
    update_notebook1()


# OBSERVE CHANGES IN THE WIDGETS, PASS THE CHANGE TO CHANGE_X OR CHANGE_Y
dropdown_feature1.observe(change_x1, names='value')
display(dropdown_feature1)
dropdown_feature2.observe(change_y1, names='value')
display(dropdown_feature2)
```

```python
from scipy.stats import pearsonr
from scipy import stats

ami = reduced_dataset['automated readability index'].values.T.tolist()
cli = reduced_dataset['coleman liau index'].values.T.tolist()
fre = reduced_dataset['flesch reading ease'].values.T.tolist()
downloads = reduced_dataset['downloads'].values.T.tolist()
rank = reduced_dataset['rank'].values.T.tolist()


def find_regular_data(data):
    outliers_x = outliers(data[0])
    outliers_y = outliers(data[1])
    data = separate_data(data, outliers_x, outliers_y)
    return data[1]

ami_cli_reg = find_regular_data([ami, cli])
ami_fre_reg = find_regular_data([ami, fre])
cli_fre_reg = find_regular_data([cli, fre])
download_rank_reg = find_regular_data([downloads, rank])


ami_cli = pearsonr(ami_cli_reg[0], ami_cli_reg[1])
ami_fre = pearsonr(ami_fre_reg[0], ami_fre_reg[1])
cli_fre = pearsonr(cli_fre_reg[0], cli_fre_reg[1])
download_rank = pearsonr(download_rank_reg[0], download_rank_reg[1])

df = pd.DataFrame({'Comparison': ['automated readibility vs coleman liau', 'automated readability vs. flesch reading', 'coleman liau vs. flesch reading', 'downloads vs. rank'], 
                   "Pearson's R:": [ami_cli[0], ami_fre[0], cli_fre[0], download_rank[0]],
                   'P-Value':[ami_cli[1], ami_fre[1], cli_fre[1], download_rank[1]]})
df
```

<font face="courier" color="black">
Above, we can see the Pearson's R correlation coefficient for the three measures of comprehensibility as well as the download vs. rank comparison, and the corresponding P-value for that coefficient statistic. What are these, and why are they important? Pearson's R is just the statistic we use to quantify correlation. It varies between -1 and 1, and if it's positive, it means the two variables vary directly. If it's negative, then they have an inverse relationship.<br><br>

Now, why do we need the p-value then? Well, not all relationships are equal. The p-value represents *the probability that you would have found the current result if the correlation coefficient were in fact zero (which is referred to as the null hypothesis)*. So, we want to see low p-values (ideally p should be less than .05 percent).<br>
<img src="https://imgs.xkcd.com/comics/p_values_2x.png" alt="P-Values">

```python
final_state = {
    'x_change': 'automated readability index',
    'y_change': 'difficult words',
}

# FIND X AND Y DATA, PASS TO UPDATE_NOTEBOOK()
def create_plot2(state):
    x_datatype = state['x_change']
    y_datatype = state['y_change']
    x_data = df[x_datatype].values.T.tolist()
    y_data = df[y_datatype].values.T.tolist()
    return x_data, y_data

    
# GET DATA
x_data1, y_data1 = create_plot2(final_state)


source_final = ColumnDataSource(data=dict(
    x=x_data1,
    y=y_data1,
    desc=titles, 
    corr=[float('%.2f' % elem) for elem in pearsonr(x_data1, y_data1)]
))

hover = HoverTool(tooltips=[
    ("(x,y)", "($x, $y)"),
    ("title", "@desc"),
])

final_figure = figure(plot_width=400, plot_height=400, tools=[hover],
           title="Correlation Plot")

final_renderer = final_figure.circle(
         'x', 
         'y', 
         size=8, 
         color='darkmagenta',
         alpha=0.6,
         source=source_final)

final_figure.title.text = "Pearson's R and P-Value: {}".format(source_final.data['corr'])
final_figure.xaxis.axis_label = final_state['x_change']
final_figure.yaxis.axis_label = final_state['y_change']

show(final_figure, notebook_handle=True)


# PUSH CHANGES TO NOTEBOOK
def update_notebook2():
    x, y = create_plot2(final_state)
    final_renderer.data_source.data['x'] = x
    final_renderer.data_source.data['y'] = y
    corr = [float('%.2f' % elem) for elem in pearsonr(x, y)]
    final_figure.xaxis.axis_label = final_state['x_change']
    final_figure.yaxis.axis_label = final_state['y_change']
    final_figure.title.text = "Pearson's R and P-Value: {}".format(corr)
    push_notebook()

# TRACK CHANGES TO X AND Y FROM DROPDOWNS
def change_x2(change):
    global final_state
    final_state['x_change'] = change.new
    update_notebook2()
    
def change_y2(change):
    global final_state
    final_state['y_change'] = change.new
    update_notebook2()

# CREATING DROPDOWN WIDGETS
final_dropdown_x = widgets.Dropdown(
    description='Feature 1 (x-axis): ',
    options = list(df),
    value = final_state['x_change'],
)

final_dropdown_y = widgets.Dropdown(
    description = 'Feature 2 (y-axis): ',
    options = list(df),
    value = final_state['y_change'],
)

# OBSERVE CHANGES IN THE WIDGETS, PASS THE CHANGE TO CHANGE_X OR CHANGE_Y
final_dropdown_x.observe(change_x2, names='value')
display(final_dropdown_x)
final_dropdown_y.observe(change_y2, names='value')
display(final_dropdown_y)
```

<h2>*One more thing - so there's this thing called "p-hacking," please avoid it...*</h2><br><br>

Usually (when we do good, ethical research), we create a hypothesis, design an experiment to test it, and then employ statistics to check if the behavior we collected in our experiment supports or fails to support our hypothesis.<br><br>
P-hacking is a phenomenon where researchers deliberately take data they have collected and run statistics on it until it shows some meaningful results, disregarding or even failing to rely on their hypotheses. You may recall those ~*nefarious*~ researchers I spoke of back in the outliers section (if not, Ctrl+F "nefarious"), and these are some of the same sorts.<br><br>

Now, if I collect data, and I find some relationship that is statistically significant. For example, let's say I found that hours per week spent watching Netflix is positively correlated with the rate of breast cancer. WHOA! Someone has to know about this! (Quick sidenote - you should already know this is b.s. because correlation isn't causation, but I digress). Isn't it my duty to HUMANITY to report findings that are going to SAVE LIVES?? Sure! But only if the original purpose of your experiment was to figure out if that relationship existed.<br><br>

Let's say, for the sake of argument, that you were actually trying to find out if rates of breast cancer were correlated with lower performance in school. Maybe you recorded how much Netflix each subject watches as another dependent variable. And maybe, the tests you ran to validate your original hypothesis were shitty. But, you need to publish research in order to keep your job as a professor. Well, well, well. We have this other statistic that *is* statistically significant, why don't we just use that? To quote <a href="https://info.umkc.edu/drbanderson/p-hacking-and-the-problem-of-multiple-comparisons/">this article</a> from one professor's explanation of p-hacking in his personal blog:<br>
    <blockquote>"we simply do not know if the strength of the relationship found is purely an artifact of the sample, the analytical method used, or legitimate judgment calls made by the researcher.  We just don’t know.  When combined with the multiple comparison problem and if statistical power is low, the probability becomes by high that the observed finding happened just by chance."</blockquote><br><br>
So, you might have just forced American's into a panic that they're all going to get breast cancer because they like to binge watch Breaking Bad. That's a serious problem. To get a better view of p-hacking (and a bonus interactive), FiveThirtyEight released <a href="https://projects.fivethirtyeight.com/p-hacking/">this interactive</a> on p-hacking during the last presidential election.<br><br>
<br>
